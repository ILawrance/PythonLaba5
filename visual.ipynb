```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
import random as rd 
import datetime
import pandas as pd
from matplotlib.dates import DayLocator, MonthLocator


x = np.array([1,2,8])
y = x
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.plot(x, y, 'g--s')
plt.show()

```

```python
z = np.arange(-10,11,2)
a = rd.randint(-5,5)
b = rd.randint(-5,5)
v = a*z+b
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.tick_params(length=8, width=2)
plt.minorticks_on()
plt.plot(z,v, 'r--s')
plt.show()

```

```python
summ_Elements=0
count=0
for i in z:
    summ_Elements+=i
    count+=1
sred = summ_Elements/count

sred = st.mean(z)
standrtOtkl = z.std()
print(standrtOtkl)

E = rd.normalvariate(sred, standrtOtkl)
w = a*z+b+E
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.tick_params(length = 1,width = 1)
plt.minorticks_on()

plt.scatter(z,w, marker='v', c='g')

plt.show()
```

```python
plt.subplot(1,2,1) # (stroki, kolonni, mesto)
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.plot(z,v,'m')
plt.subplot(1,2,2)
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.scatter(z,w, marker='+', c='b')

plt.show()
```

```python
#Курс доллара
from matplotlib.dates import DateFormatter


Date = np.array([datetime.datetime (2022, i+1, 1) for i in range(12)])
Dollar = np.array([76.5, 77.1, 103.4, 77.9, 63.3, 57.1, 58.2, 60.4, 59.8, 61.1, 60.8, 65.8])
plt.xticks(rotation = 35)
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.plot(Date, Dollar)

Months = MonthLocator(interval=1)
Date_Fmt = DateFormatter('%y-%m-%d')
plt.gca().xaxis.set_major_locator(Months)
plt.gca().xaxis.set_major_formatter(Date_Fmt)

plt.title('Курс доллара к рублю за 2022 год')
plt.xlabel('Дата')
plt.ylabel('Рубли за 1 доллар')
plt.show() 
```
