```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
import random as rd

x = np.array([1,2,8])
y = x
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.plot(x, y, 'g--s')
plt.show()

```

```python
z = np.arange(2,21,2)
a = rd.randint(1,5)
b = rd.randint(1,5)
v = a*z+b
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.plot(z,v, 'r--s')
plt.show()

```

```python
summ_Elements=0
count=0

for i in z:
    summ_Elements+=i
    count+=1

sred = summ_Elements/count
sred = st.mean(z)
standrtOtkl = z.std()
print(standrtOtkl)
E = rd.normalvariate(sred, standrtOtkl)
w = a*z+b+E
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.tick_params(length = 1,width = 1)
plt.minorticks_on()

plt.scatter(z,w, marker='v', c='g')

plt.show()
```

```python
plt.subplot(1,2,1) # (stroki, kolonni, mesto)
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.plot(z,v,'m')
plt.subplot(1,2,2)
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.scatter(z,w, marker='+', c='b')

plt.show()
```
