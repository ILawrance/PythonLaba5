```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
import random as rd 
import datetime
import pandas as pd
from matplotlib.dates import MonthLocator
from matplotlib.dates import DateFormatter

axis_X = np.array([1,2,8])
axis_Y = axis_X
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.plot(axis_X, axis_Y, 'g--s')
plt.show()

```

```python
axis_X = np.arange(-10,11,2)
factor_A = rd.randint(1,10)
factor_B = rd.randint(1,10)
axis_Y = factor_A * axis_X + factor_B
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.tick_params(length=8, width=2)
plt.minorticks_on()
plt.plot(axis_X, axis_Y, 'r--s')
plt.show()

```

```python
standrtOtkl = axis_X.std()
print(standrtOtkl)

factor_E = np.random.normal(0, standrtOtkl, 11)
axis_Y2 = factor_A * axis_X + factor_B + factor_E
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.tick_params(length = 1,width = 1)
plt.minorticks_on()
plt.scatter(axis_X, axis_Y2, marker='v', c='g')
plt.plot(axis_X, axis_Y, 'r--s')
plt.show()
```

```python
plt.subplot(1,2,1) # (stroki, kolonni, mesto)
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.plot(axis_X, axis_Y,'m')
plt.subplot(1,2,2)
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.scatter(axis_X, axis_Y2, marker='+', c='b')

plt.show()
```

```python
#Курс доллара
Date = np.array([datetime.datetime (2022, i+1, 1) for i in range(12)])
Dollar = np.array([76.5, 77.1, 103.4, 77.9, 63.3, 57.1, 58.2, 60.4, 59.8, 61.1, 60.8, 65.8])
plt.xticks(rotation = 35)
plt.grid(linewidth = 1, color = 'gray', linestyle = '--')
plt.minorticks_on()
plt.plot(Date, Dollar)

Months = MonthLocator(interval=1)
Date_Fmt = DateFormatter('%y-%m-%d')
plt.gca().xaxis.set_major_locator(Months)
plt.gca().xaxis.set_major_formatter(Date_Fmt)

plt.title('Курс доллара к рублю за 2022 год')
plt.xlabel('Дата')
plt.ylabel('Рубли за 1 доллар')
plt.show() 
```
